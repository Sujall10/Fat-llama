```python
churn_df = updated_bookings.groupby('user_id').agg(
    total_bookings=('booking_id', 'count'),
    last_booking_date=('booking_date', 'max'),
    first_booking_date=('booking_date', 'min'),
    total_spent=('final_payment', 'sum'),
    avg_booking_fee=('booking_fee', 'mean'),
    avg_final_payment=('final_payment', 'mean'),
    avg_rental_days=('rental_days', 'mean'),
    avg_rental_duration_hrs=('rental_duration', 'mean'),
    avg_price_per_day=('price_per_day', 'mean'),
    avg_price_per_hour=('price_per_hour', 'mean'),
    avg_rating=('rating', 'mean'),
    review_count=('review', lambda x: x.notna().sum()),
    unique_items=('item_id', pd.Series.nunique),
    unique_categories=('category', pd.Series.nunique),
    unique_subcategories=('subcategory', pd.Series.nunique),
    unique_providers=('provider_id', pd.Series.nunique),
    unique_locations=('location_id', pd.Series.nunique),
    avg_payment_ratio=('payment_ratio', 'mean'),
    avg_expected_payment=('expected_payment', 'mean')
).reset_index()
churn_df
```

```python
churn_df['review_rate'] = churn_df['review_count'] / churn_df['total_bookings']
churn_df['days_active'] = (churn_df['last_booking_date'] - churn_df['first_booking_date']).dt.days
churn_df['booking_frequency_per_month'] = churn_df['total_bookings'] / (churn_df['days_active'] / 30 + 0.01)
churn_df
```

```python
updated_bookings['is_negative_review'] = updated_bookings['rating'] <= 2

negative_reviews_df = updated_bookings.groupby('user_id').agg(
    negative_review_count=('is_negative_review', 'sum'),
    negative_review_ratio=('is_negative_review', 'mean')
).reset_index()
churn_df = churn_df.merge(negative_reviews_df, on='user_id', how='left')
```

```python
churn_df
```

```python
# Add a column to define early vs late phase (e.g., based on median date per user)
def tag_booking_phase(df):
    df = df.sort_values('booking_date')
    midpoint = df['booking_date'].median()
    df['phase'] = np.where(df['booking_date'] <= midpoint, 'early', 'late')
    return df

# Apply per user
past_bookings = updated_bookings.groupby('user_id').apply(tag_booking_phase).reset_index(drop=True)

# Count bookings in each phase
freq_phase_df = past_bookings.groupby(['user_id', 'phase']).agg(
    bookings=('booking_id', 'count'),
    days_active=('booking_date', lambda x: (x.max() - x.min()).days + 1)
).reset_index()

# Pivot into wide format
freq_wide = freq_phase_df.pivot(index='user_id', columns='phase', values=['bookings', 'days_active'])
freq_wide.columns = ['_'.join(col).strip() for col in freq_wide.columns.values]

# Calculate frequency per day
freq_wide['early_freq'] = freq_wide['bookings_early'] / freq_wide['days_active_early']
freq_wide['late_freq'] = freq_wide['bookings_late'] / freq_wide['days_active_late']

# Calculate change rate
freq_wide['frequency_change_rate'] = (freq_wide['late_freq'] - freq_wide['early_freq']) / (freq_wide['early_freq'] + 0.01)

churn_df = churn_df.merge(freq_wide[['frequency_change_rate']], left_on='user_id', right_index=True, how='left')

churn_df
```
